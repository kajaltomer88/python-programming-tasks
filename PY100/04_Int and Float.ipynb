# Common Handy Integer and Float Functions
### Official documentation: https://docs.python.org/3/library/functions.html

## Core Built-in Functions
 - `abs(x)`: Returns the absolute value (e.g., `abs(-5)` → `5`).
 - `round(number, ndigits)`: Rounds a float to a specified number of decimal places. If `ndigits` is omitted, returns the nearest integer.
 - `pow(base, exp)`: Returns `base ** exp`. With a third argument `pow(base, exp, mod)` it computes modular exponentiation efficiently.
 - `divmod(a, b)`: Returns a tuple `(a // b, a % b)` — the quotient and remainder in one call.

## Type Conversion
 - `int(x)`: Converts a float or numeric string to an integer (truncates toward zero).
 - `float(x)`: Converts an integer or numeric string to a float.

## Useful `math` Module Functions (import math)
 - `math.floor(x)` / `math.ceil(x)`: Round down / round up to the nearest integer.
 - `math.sqrt(x)`: Square root.
 - `math.pi`, `math.e`: Mathematical constants.

> **Perspective:** Python's `int` has **unlimited precision** — it can hold numbers as large as your memory allows. This is different from C/Java where integers overflow at fixed bit widths.

```python
intvar = -360520
floatvar = 3.14159

print(intvar)
print(abs(intvar))           # Absolute value

print(round(intvar, 3))      # No effect — integer has no decimals
print(floatvar)
print(round(floatvar, 2))    # Rounds to 2 decimal places → 3.14

print(divmod(intvar, 100))   # Quotient and remainder as a tuple
```

```python
## Type Conversion and the `math` Module in Action

import math

print(int(3.9))              # Truncates toward zero → 3
print(float(7))              # Converts int to float → 7.0

print(math.floor(3.7))      # Always rounds down → 3
print(math.ceil(3.2))       # Always rounds up → 4
print(math.sqrt(144))       # Square root → 12.0

print(f"Pi = {math.pi}")
print(f"e  = {math.e}")

# pow() with modular arithmetic: (2^10) % 1000
print(pow(2, 10, 1000))     # → 24
```
