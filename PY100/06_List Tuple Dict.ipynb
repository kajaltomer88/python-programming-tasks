# Python Collections — List, Tuple, and Dictionary
### Official documentation: https://docs.python.org/3/library/stdtypes.html

Python provides several built-in collection types. The three most commonly used are:

| Collection | Syntax | Ordered | Mutable | Duplicates | Use Case |
|---|---|---|---|---|---|
| **List** | `[ ]` | Yes | Yes | Yes | General-purpose ordered collection |
| **Tuple** | `( )` | Yes | **No** | Yes | Fixed data, dictionary keys, return values |
| **Dictionary** | `{ }` | Yes (3.7+) | Yes | Keys: No, Values: Yes | Key-value lookups |

> **Perspective:** Choosing the right collection type is a fundamental design decision. Use a **List** when order matters and data changes, a **Tuple** when data should not change, and a **Dictionary** when you need fast lookups by key.

---
## 1. List — Ordered, Mutable Collection

A `list` is the most versatile collection. It can hold mixed types (though keeping a single type is usually cleaner).

```python
## Creating Lists

fruits = ["apple", "banana", "cherry"]
numbers = [10, 20, 30, 40, 50]
mixed = [1, "hello", 3.14, True]       # Allowed but not recommended

print(f"Fruits  : {fruits}")
print(f"Numbers : {numbers}")
print(f"Mixed   : {mixed}")
print(f"Length  : {len(fruits)}")
```

```python
## Accessing and Slicing

fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[0])        # First element → apple
print(fruits[-1])       # Last element → elderberry
print(fruits[1:3])      # Slice from index 1 to 2 → ['banana', 'cherry']
print(fruits[:3])       # First 3 elements → ['apple', 'banana', 'cherry']
print(fruits[2:])       # From index 2 to end → ['cherry', 'date', 'elderberry']
print(fruits[::2])      # Every 2nd element → ['apple', 'cherry', 'elderberry']
print(fruits[::-1])     # Reversed → ['elderberry', 'date', 'cherry', 'banana', 'apple']
```

```python
## Modifying Lists

fruits = ["apple", "banana", "cherry"]

# Add elements
fruits.append("date")             # Add to end
fruits.insert(1, "avocado")       # Insert at index 1
print(f"After add   : {fruits}")

# Remove elements
fruits.remove("banana")           # Remove by value (first occurrence)
popped = fruits.pop()             # Remove and return last element
print(f"Popped      : {popped}")
print(f"After remove: {fruits}")

# Update an element
fruits[0] = "apricot"             # Replace by index
print(f"After update: {fruits}")
```

```python
## Useful List Methods

numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"Count of 1  : {numbers.count(1)}")
print(f"Index of 5  : {numbers.index(5)}")

numbers.sort()                     # Sort in place (ascending)
print(f"Sorted      : {numbers}")

numbers.sort(reverse=True)         # Sort descending
print(f"Descending  : {numbers}")

numbers.reverse()                  # Reverse the current order
print(f"Reversed    : {numbers}")

# sorted() returns a NEW list — does not modify the original
original = [3, 1, 2]
new_sorted = sorted(original)
print(f"Original    : {original}")
print(f"New sorted  : {new_sorted}")
```

```python
## List Comprehension — concise way to create lists

# Squares of 0–9
squares = [x**2 for x in range(10)]
print(f"Squares     : {squares}")

# Filter: only even numbers
evens = [x for x in range(20) if x % 2 == 0]
print(f"Evens       : {evens}")

# Transform: uppercase a list of strings
names = ["alice", "bob", "charlie"]
upper_names = [name.upper() for name in names]
print(f"Uppercased  : {upper_names}")
```

## 2. Tuple — Ordered, Immutable Collection

A `tuple` is like a list that **cannot be changed** after creation. This makes it:
- Safer for data that should not be modified
- Usable as dictionary keys (lists cannot be used as keys)
- Slightly faster than lists for iteration

```python
## Creating Tuples

coordinates = (10.0, 20.0)
rgb_color = (255, 128, 0)
single = (42,)                     # Note the trailing comma — without it, (42) is just an int
from_list = tuple([1, 2, 3])       # Convert a list to a tuple

print(f"Coordinates : {coordinates}")
print(f"RGB         : {rgb_color}")
print(f"Single      : {single}, type: {type(single)}")
print(f"From list   : {from_list}")
```

```python
## Accessing and Unpacking Tuples

point = (10, 20, 30)

# Access by index
print(f"First  : {point[0]}")
print(f"Last   : {point[-1]}")

# Unpacking — assign each element to a variable
x, y, z = point
print(f"x={x}, y={y}, z={z}")

# Unpacking with * (star) to capture remaining elements
first, *rest = (1, 2, 3, 4, 5)
print(f"First: {first}, Rest: {rest}")

# Tuples are immutable — this would raise TypeError:
# point[0] = 99  # ❌ TypeError: 'tuple' object does not support item assignment
```

```python
## Tuple Methods and Operations

data = (1, 2, 3, 2, 4, 2)

print(f"Count of 2   : {data.count(2)}")
print(f"Index of 3   : {data.index(3)}")
print(f"Length        : {len(data)}")
print(f"2 in data?   : {2 in data}")

# Concatenation creates a NEW tuple
combined = (1, 2) + (3, 4)
print(f"Combined     : {combined}")

# Common use: returning multiple values from a function
def min_max(numbers):
    return min(numbers), max(numbers)   # Returns a tuple

lo, hi = min_max([5, 2, 8, 1, 9])
print(f"Min={lo}, Max={hi}")
```

## 3. Dictionary — Key-Value Pairs

A `dict` stores data as **key: value** pairs. Keys must be unique and immutable (strings, numbers, tuples). Values can be anything.

```python
## Creating Dictionaries

student = {
    "name": "Shri Prakash",
    "age": 25,
    "course": "AI Internship",
    "active": True
}

print(f"Student     : {student}")
print(f"Name        : {student['name']}")
print(f"Keys        : {list(student.keys())}")
print(f"Values      : {list(student.values())}")
print(f"Items       : {list(student.items())}")
```

```python
## Accessing Values — [] vs .get()

student = {"name": "Shri Prakash", "age": 25}

# Direct access — raises KeyError if key is missing
print(student["name"])

# Safe access — returns None (or a default) if key is missing
print(student.get("email"))             # None
print(student.get("email", "N/A"))      # "N/A" (default value)

# Tip: Use .get() when you are not sure if the key exists
```

```python
## Modifying Dictionaries

student = {"name": "Shri Prakash", "age": 25}

# Add or update
student["email"] = "shri@example.com"   # Add new key
student["age"] = 26                     # Update existing key
print(f"Updated : {student}")

# Remove
del student["email"]                    # Delete by key
removed = student.pop("age")            # Remove and return value
print(f"Removed age: {removed}")
print(f"After remove: {student}")

# Merge two dictionaries
defaults = {"theme": "dark", "lang": "en"}
overrides = {"lang": "hi", "font": "mono"}
merged = {**defaults, **overrides}      # Spread operator — overrides win
print(f"Merged  : {merged}")
```

```python
## Iterating Over a Dictionary

scores = {"Alice": 90, "Bob": 85, "Charlie": 92}

# Iterate over keys (default)
for name in scores:
    print(f"{name} scored {scores[name]}")

print("---")

# Iterate over key-value pairs (preferred)
for name, score in scores.items():
    print(f"{name} → {score}")

print("---")

# Dictionary Comprehension
doubled = {name: score * 2 for name, score in scores.items()}
print(f"Doubled : {doubled}")
```

## Quick Comparison: When to Use What?

| Need | Use |
|---|---|
| Ordered collection that changes | **List** |
| Fixed data, function return values | **Tuple** |
| Fast lookup by key | **Dictionary** |
| Unique items only | **Set** `{ }` (bonus — covered in a later lesson) |

> **Key Insight:** Lists and Dicts are passed by **reference** in Python. If you pass a list to a function and modify it inside, the original list is also modified. Use `.copy()` or `list()` to create a shallow copy when you need independence.
