## Compiler vs Interpreter (Is Python a Compiled or Interpreted Language?)

### Compiler: [C, C++, Rust, Go]
    A compiler translates your entire source code into machine code (an executable) in one go.
    
    **Advantages:**
    1. Faster runtime performance
    2. Better optimization
    3. Easier to ship (single binary)
    4. Good for system-level work (low-level control, memory, hardware, etc.)

    **Disadvantages:**
    1. Slower development cycle
    2. Platform-specific builds
    3. Harder to debug
    4. Heavier tooling   

### Interpreter: [Python, JavaScript, PHP, Ruby]
    An interpreter runs code line by line via a virtual machine.

    **Advantages:**
    1. Faster iteration / easier to experiment
    2. Great for scripting and automation
    3. More portable across platforms
    4. Friendlier debugging

    **Disadvantages:**
    1. Slower runtime
    2. Requires a runtime / VM
    3. Fewer optimization opportunities

> **Note:** Java is often called "interpreted" but it actually compiles to bytecode first, then runs on the JVM — making it a hybrid approach. Python similarly compiles to `.pyc` bytecode before interpreting it.

## Static (Early Binding) vs Dynamic (Runtime Binding)

### Static Binding — resolved at compile time (before running)
- Faster calls, more compile-time error checking, easier for the compiler to optimize
- Less flexible behaviour

### Dynamic Binding — resolved at runtime based on context (actual object involved)
- Very flexible (polymorphism, "duck typing"); enables plugin architectures and runtime extensions
- Runtime overhead, more prone to logical errors

### Python uses Dynamic Binding for assignment — variables, functions, and types need no separate declaration.

> **"With greater freedom comes greater responsibility."** — Always test thoroughly when relying on dynamic behaviour.

## Error Types

### 1. SyntaxError — caught before the code runs
A `SyntaxError` occurs when Python cannot parse your code. For example, a missing parenthesis or an incorrect keyword.
- **IndentationError** is a specific kind of `SyntaxError` that occurs when indentation is inconsistent.

### 2. Exception / Runtime Error — occurs during execution
These errors happen while the program is running. Common causes include invalid input, division by zero, or accessing a non-existent file.
- **TypeError:** Performing an operation on an incompatible data type (e.g., `"hello" + 5`).
- **ValueError:** Right type but wrong value (e.g., `int("abc")`).
- **FileNotFoundError:** Trying to open a file that does not exist.

### 3. Logical Error — the code runs but produces wrong results
The program executes without any error messages, yet the output is incorrect. For example, calculating an average but forgetting to divide by the count.

> **Tip:** Logical errors are the hardest to find because Python won't warn you. Writing unit tests and using `print()` / debugger statements are your best friends here.
