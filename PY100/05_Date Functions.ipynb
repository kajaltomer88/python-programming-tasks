# Common Handy Date and Time Functions
### Official documentation: https://docs.python.org/3/library/datetime.html

Python provides two key modules for working with dates and times:
- **`datetime`** — classes for manipulating dates, times, and timestamps.
- **`time`** — lower-level time functions (sleep, performance counters, etc.).

## Key Classes in `datetime`
| Class | Purpose |
|---|---|
| `date` | Year, month, day only |
| `time` | Hour, minute, second, microsecond only |
| `datetime` | Combined date and time |
| `timedelta` | A duration / difference between two dates or times |

> **Perspective:** Always store and compute dates using `datetime` objects — not strings. Convert to strings only for display. This avoids countless bugs around formatting and comparison.

```python
## Getting the Current Date and Time

from datetime import datetime, date, time, timedelta

now = datetime.now()
print(f"Current date and time : {now}")
print(f"Date only            : {now.date()}")
print(f"Time only            : {now.time()}")
print(f"Year: {now.year}, Month: {now.month}, Day: {now.day}")
print(f"Hour: {now.hour}, Minute: {now.minute}, Second: {now.second}")
```

## Formatting Dates — `strftime()` (String Format Time)

Convert a `datetime` object into a human-readable string using format codes:

| Code | Meaning | Example |
|---|---|---|
| `%Y` | 4-digit year | 2026 |
| `%m` | Zero-padded month | 02 |
| `%d` | Zero-padded day | 21 |
| `%H` | Hour (24-hour) | 14 |
| `%I` | Hour (12-hour) | 02 |
| `%M` | Minute | 30 |
| `%S` | Second | 45 |
| `%A` | Full weekday name | Friday |
| `%B` | Full month name | February |
| `%p` | AM/PM | PM |

```python
## strftime() — datetime to string

now = datetime.now()

print(now.strftime("%Y-%m-%d"))           # 2026-02-21 (ISO format)
print(now.strftime("%d/%m/%Y"))           # 21/02/2026 (DD/MM/YYYY)
print(now.strftime("%A, %B %d, %Y"))     # Friday, February 21, 2026
print(now.strftime("%I:%M %p"))           # 02:30 PM (12-hour format)
```

```python
## strptime() — string to datetime (parsing)

date_string = "2026-02-21"
parsed = datetime.strptime(date_string, "%Y-%m-%d")
print(f"Parsed: {parsed}")
print(f"Type  : {type(parsed)}")

# A common real-world pattern: parse a date from user input, then work with it as datetime
user_input = "21/02/2026"
parsed2 = datetime.strptime(user_input, "%d/%m/%Y")
print(f"Parsed from DD/MM/YYYY: {parsed2}")
```

## `timedelta` — Date Arithmetic

A `timedelta` represents a **duration** — the difference between two dates or times. You can add or subtract them from `datetime` objects.

```python
## timedelta — adding and subtracting dates

today = date.today()
print(f"Today       : {today}")

# Add 30 days
future = today + timedelta(days=30)
print(f"In 30 days  : {future}")

# Subtract 7 days
past = today - timedelta(days=7)
print(f"7 days ago  : {past}")

# Difference between two dates
new_year = date(2027, 1, 1)
diff = new_year - today
print(f"Days until 2027: {diff.days}")
print(f"Type of diff   : {type(diff)}")
```

```python
## Comparing Dates

date1 = date(2026, 1, 1)
date2 = date(2026, 12, 31)

print(f"date1 < date2 : {date1 < date2}")    # True — Jan comes before Dec
print(f"date1 == date2: {date1 == date2}")    # False
print(f"date1 > date2 : {date1 > date2}")    # False

# Practical example: checking if a deadline has passed
deadline = date(2026, 3, 1)
if date.today() > deadline:
    print("Deadline has passed!")
else:
    print(f"You have {(deadline - date.today()).days} days left.")
```

```python
## Replacing Parts of a Date

now = datetime.now()
print(f"Original : {now}")

# Replace the year
modified = now.replace(year=2030)
print(f"Year=2030: {modified}")

# Replace month and day
modified2 = now.replace(month=1, day=1)
print(f"Jan 1    : {modified2}")

# Note: replace() returns a NEW datetime — the original is unchanged (immutable)
```

## Quick Reference: `strftime` vs `strptime`

| Method | Direction | Mnemonic |
|---|---|---|
| `strftime()` | datetime → string | **f** = **f**ormat |
| `strptime()` | string → datetime | **p** = **p**arse |

> **Tip:** When working with time zones, consider using the `zoneinfo` module (Python 3.9+) or the third-party `pytz` library. Naive datetimes (without timezone info) can cause subtle bugs when comparing times across regions.
