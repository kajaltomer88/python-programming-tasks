## 1. `for` with `range()` — Repeat a Fixed Number of Times

`range(start, stop, step)` generates a sequence of numbers.
- `start` defaults to `0`
- `stop` is **exclusive** (not included)
- `step` defaults to `1`

```python
## range() — basic usage

# Print 0 through 4
for i in range(5):
    print(i, end=" ")
print()  # newline

# range(start, stop) — from 2 to 6
for i in range(2, 7):
    print(i, end=" ")
print()

# range(start, stop, step) — count by 3s
for i in range(0, 20, 3):
    print(i, end=" ")
print()

# Counting backwards
for i in range(10, 0, -2):
    print(i, end=" ")
print()
```

## 2. `for` Over a List

The most common use of `for` — iterate over every element in a list (or any iterable).

```python
## Iterating over a list

fruits = ["apple", "banana", "cherry", "date"]

for fruit in fruits:
    print(f"I like {fruit}")

print("---")

# Iterating over a string (strings are iterable too!)
for char in "Python":
    print(char, end="-")
print()

# Summing a list of numbers
prices = [10.5, 20.0, 35.75, 8.25]
total = 0
for price in prices:
    total += price
print(f"Total: ${total:.2f}")

# Tip: Python's built-in sum() does this in one call
print(f"Total (built-in): ${sum(prices):.2f}")
```

## 3. `enumerate()` — Get Index and Value Together

Instead of using `range(len(list))` to track the index, use `enumerate()`. It is more Pythonic and less error-prone.

```python
## enumerate() — the Pythonic way to get index + value

languages = ["Python", "JavaScript", "Rust", "Go"]

# The old way (avoid this)
print("--- Using range(len()) ---")
for i in range(len(languages)):
    print(f"{i}: {languages[i]}")

# The Pythonic way
print("--- Using enumerate() ---")
for index, lang in enumerate(languages):
    print(f"{index}: {lang}")

# Start counting from 1 instead of 0
print("--- Starting from 1 ---")
for rank, lang in enumerate(languages, start=1):
    print(f"#{rank}: {lang}")
```

## 4. `for` Over a Dictionary

Dictionaries offer multiple ways to iterate: over keys, values, or key-value pairs.

```python
## Iterating over a dictionary

scores = {"Alice": 90, "Bob": 85, "Charlie": 92, "Diana": 88}

# Iterate over keys (default)
print("--- Keys ---")
for name in scores:
    print(name)

# Iterate over values
print("--- Values ---")
for score in scores.values():
    print(score)

# Iterate over key-value pairs (most useful)
print("--- Items ---")
for name, score in scores.items():
    status = "Pass" if score >= 60 else "Fail"
    print(f"{name}: {score} ({status})")

# Find the top scorer
top_name = max(scores, key=scores.get)
print(f"\nTop scorer: {top_name} with {scores[top_name]}")
```

## 5. `while` Loop — Repeat Until a Condition Changes

A `while` loop keeps running as long as its condition is `True`. Be careful — if the condition never becomes `False`, you get an **infinite loop**.

```python
## while loop — basic countdown

count = 5
while count > 0:
    print(f"Countdown: {count}")
    count -= 1          # Don't forget to update the condition!
print("Liftoff!")

print("---")

# Practical example: keep doubling until we exceed 1000
value = 1
steps = 0
while value <= 1000:
    value *= 2
    steps += 1
print(f"Reached {value} in {steps} doublings")
```

## 6. `break`, `continue`, and `else` on Loops

| Keyword | Effect |
|---|---|
| `break` | Exit the loop immediately |
| `continue` | Skip the rest of the current iteration, go to the next one |
| `else` (on loop) | Runs only if the loop completed **without** hitting `break` |

```python
## break — exit the loop early

# Find the first negative number
numbers = [10, 25, 3, -5, 8, 14]

for num in numbers:
    if num < 0:
        print(f"Found negative number: {num}")
        break
    print(f"Checking: {num}")

# After break, execution continues here (outside the loop)
```

```python
## continue — skip the current iteration

# Print only odd numbers
for i in range(10):
    if i % 2 == 0:
        continue          # Skip even numbers
    print(i, end=" ")
print()
```

```python
## else on a loop — runs only if no break occurred

# Searching for a value
target = 7
numbers = [1, 3, 5, 7, 9]

for num in numbers:
    if num == target:
        print(f"Found {target}!")
        break
else:
    # This runs ONLY if the loop finished without break
    print(f"{target} not found in the list.")

print("---")

# Try with a value that does not exist
target = 4
for num in numbers:
    if num == target:
        print(f"Found {target}!")
        break
else:
    print(f"{target} not found in the list.")
```

```python
## while with break — input validation pattern

# Simulate user input validation (using a list as mock input)
mock_inputs = ["", "  ", "hello"]

for user_input in mock_inputs:
    stripped = user_input.strip()
    if stripped:
        print(f"Valid input received: '{stripped}'")
        break
    else:
        print(f"Empty input '{user_input}' — try again")
```

## 7. Nested Loops

A loop inside a loop. The inner loop runs completely for each iteration of the outer loop.

```python
## Nested loops — multiplication table

for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i*j:4}", end="")
    print()  # newline after each row

print("---")

# Practical example: finding pairs that sum to a target
numbers = [1, 3, 5, 7, 9]
target_sum = 10

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):   # Start from i+1 to avoid duplicates
        if numbers[i] + numbers[j] == target_sum:
            print(f"{numbers[i]} + {numbers[j]} = {target_sum}")
```

## 8. `zip()` — Iterate Over Multiple Sequences in Parallel

`zip()` pairs up elements from two (or more) iterables. It stops when the **shortest** iterable is exhausted.

```python
## zip() — parallel iteration

names = ["Alice", "Bob", "Charlie"]
scores = [90, 85, 92]

for name, score in zip(names, scores):
    print(f"{name}: {score}")

print("---")

# Zip three lists
subjects = ["Math", "Science", "English"]
for name, subj, score in zip(names, subjects, scores):
    print(f"{name} got {score} in {subj}")

print("---")

# Creating a dictionary from two lists using zip
student_scores = dict(zip(names, scores))
print(f"Dictionary: {student_scores}")
```

---
## Quick Reference

| Pattern | When to Use |
|---|---|
| `for i in range(n)` | Repeat n times (counter-based) |
| `for item in collection` | Process each element |
| `for i, item in enumerate(collection)` | Need both index and element |
| `for key, value in dict.items()` | Iterate over dictionary entries |
| `for a, b in zip(list1, list2)` | Iterate two lists in parallel |
| `while condition` | Repeat until a condition changes |
| `break` | Exit a loop early |
| `continue` | Skip to the next iteration |

> **Perspective:** Prefer `for` loops over `while` loops whenever the number of iterations is known or you are iterating over a collection. `while` loops are best when you are waiting for a condition to change (user input, convergence, external event). Also, consider **list comprehensions** (covered in notebook 06) as a more concise alternative to simple `for` loops that build a list.
