# Common Handy String Functions
### Official documentation: https://docs.python.org/3/library/stdtypes.html#string-methods

## Case Conversion
 - `.lower()`: Converts all characters to lowercase.
 - `.upper()`: Converts all characters to uppercase.
 - `.capitalize()`: Uppercases the first character; lowercases the rest.
 - `.title()`: Uppercases the first character of each word.

## Whitespace and Character Removal
  - `.strip()`, `.lstrip()`, `.rstrip()`: Remove leading/trailing characters (defaults to whitespace).

## Searching and Checking
  - `.find(sub)` / `.index(sub)`: Locate the first occurrence of a substring. (`find` returns `-1` if not found; `index` raises a `ValueError`.)
  - `.count(sub)`: Counts how many times a substring appears.
  - `.startswith(prefix)` / `.endswith(suffix)`: Check if a string begins or ends with a given substring.

## Other Common Operations
  - `.split(separator)`: Breaks a string into a **list** based on a delimiter (default is whitespace).
  - `.join(iterable)`: Concatenates elements of an iterable into a single string using the calling string as separator.
  - `.replace(old, new)`: Returns a copy of the string with all occurrences of `old` replaced by `new`.

## Validation
 - `.isalnum()`: `True` if all characters are alphanumeric (letters or numbers).
 - `.isalpha()`: `True` if all characters are alphabetic.
 - `.isdigit()`: `True` if all characters are digits.
 - `.isspace()`: `True` if the string consists only of whitespace.
 - `.islower()` / `.isupper()`: `True` if all cased characters are lower or upper.

> **Perspective:** Strings in Python are **immutable** — every method above returns a *new* string. The original is never modified. This is a key concept that distinguishes strings from lists.

```python
myname = "        shri praakash          singh            "
print(myname)
print(myname.lstrip())              # Remove leading whitespace
print(myname.rstrip())              # Remove trailing whitespace
print(myname.strip())               # Remove both leading and trailing whitespace

print(myname.strip().upper())       # UPPERCASE
print(myname.strip().lower())       # lowercase
print(myname.strip().title())       # Title Case
print(myname.strip().capitalize())  # Capitalize first letter only

print(myname.strip().replace(" ", "."))   # Replace spaces with dots
print(myname.strip().split())             # Split into a list of words
print(myname.strip().split()[0])          # First word (index 0)
print(myname.strip().split()[2])          # Third word (index 2)

# All sequences in Python are zero-indexed: first element is at index 0,
# second at index 1, and so on. So the third element uses index 2.
```
