# PY100 - `print()` — Your Most Handy Debugging and Output Function

### Official documentation: https://docs.python.org/3/library/functions.html#print
## `print(*objects, sep=' ', end='\n', file=None, flush=False)`

- `print()` can take **any number of objects**; all non-keyword arguments are converted to strings (like `str()`).
- Objects are **separated by a space** (`sep=' '`) by default.
- Each `print()` call **ends with a newline** (`end='\n'`) by default.

> **Perspective:** `print()` is not just for output — it is often the quickest way to debug. Sprinkle `print()` calls to inspect variable values before reaching for a full debugger.

```python
## Example: Printing a string 

name = "Shri Prakash Singh"
print(name)
print("Welcome to AI Internship", "Program of 2026")

# Prints name, then on the next line notice the space between "Internship" and "Program" — that's the default sep=' '
```

```python
## Changing the end character — removing the default newline

name = "Shri Prakash Singh"
print("Hello, my name is", name, "I am your instructor and", end=" ")
print("Welcome to AI Internship", "Program of 2026")

# Both print() calls appear on a SINGLE line because end=" " replaced the newline with a space
```

## Parameters: Positional vs Named (Keyword)

### Positional parameters are resolved by the order they are passed to the function.
For example, the first value in `print()` is printed first, followed by the next, separated by `sep`.

### Named (Keyword) parameters are specified by name, like `end=`, `sep=`, `file=`, etc.
They can appear in any order and override default behaviour.

> **Key Insight:** In your own functions, you can mix both: positional args come first, keyword args come after. This is a core Python pattern you will see everywhere.

## Formatted String Printing (f-strings)

We can tell Python to treat a string as a mix of literal text and variable values by:
1. Prefixing the string with `f`
2. Placing variable names (or expressions) inside `{}`

> **f-strings** were introduced in Python 3.6 and are the recommended way to format strings. They are faster and more readable than `.format()` or `%` formatting.

```python
# Without f-string: {name} is printed literally — Python does not substitute the variable
print("Hello, my name is {name}, I am your instructor.")

# With f-string: Python evaluates {name} and inserts the value
print(f"Hello, my name is {name}, I am your instructor.")

# You can also put expressions inside the braces
print(f"2 + 2 = {2 + 2}")
```
