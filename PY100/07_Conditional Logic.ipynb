# Conditional Logic in Python
### Official documentation: https://docs.python.org/3/tutorial/controlflow.html

Conditional statements let your program make **decisions** — execute different blocks of code based on whether a condition is `True` or `False`.

## Tools Covered in This Notebook
1. `if` — single condition
2. `if...else` — two branches
3. `elif` — multiple branches
4. `match...case` — structural pattern matching (Python 3.10+)
5. Ternary operator — one-line conditional expression

> **Perspective:** Every program is essentially *data + decisions*. Mastering conditional logic is what separates writing scripts from writing real software.

---
## 1. `if` — Execute Only When True

The simplest conditional. The indented block runs only if the condition evaluates to `True`.

```python
## if — basic condition check

temperature = 38

if temperature > 37:
    print("You have a fever. Please rest!")

# Nothing prints if temperature <= 37 — the block is simply skipped

age = 20
if age >= 18:
    print("You are eligible to vote.")
```

---
## 2. `if...else` — Two Branches

When you need to do one thing if `True` and another if `False`.

```python
## if...else — two possible outcomes

number = 15

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")

# Checking login status
is_logged_in = False

if is_logged_in:
    print("Welcome back!")
else:
    print("Please log in to continue.")
```

---
## 3. `elif` — Multiple Branches

When there are more than two possibilities, use `elif` (short for "else if"). Python checks each condition from top to bottom and executes the **first** one that is `True`.

```python
## elif — grading system

score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
```

```python
## Combining conditions with `and`, `or`, `not`

age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive.")

if age < 13 or age > 65:
    print("You get a discount.")
else:
    print("Standard pricing applies.")

is_blocked = False
if not is_blocked:
    print("Access granted.")
```

## 4. `match...case` — Structural Pattern Matching (Python 3.10+)

`match` is Python's answer to `switch` in other languages, but far more powerful. It matches **patterns**, not just values.

> **Note:** This requires Python 3.10 or later. If you are on an older version, use `if/elif/else` instead.

```python
## match...case — basic value matching

command = "start"

match command:
    case "start":
        print("Starting the application...")
    case "stop":
        print("Stopping the application...")
    case "restart":
        print("Restarting the application...")
    case _:                          # _ is the wildcard / default case
        print(f"Unknown command: {command}")
```

```python
## match...case — matching with patterns and guards

# Matching HTTP status codes
status_code = 404

match status_code:
    case 200:
        print("OK")
    case 301 | 302:                  # OR pattern — matches either value
        print("Redirect")
    case 404:
        print("Not Found")
    case 500:
        print("Internal Server Error")
    case code if 400 <= code < 500:  # Guard clause — additional condition
        print(f"Client Error: {code}")
    case _:
        print(f"Other status: {status_code}")
```

```python
## match...case — matching data structures (the real power)

point = (0, 5)

match point:
    case (0, 0):
        print("Origin")
    case (x, 0):
        print(f"On the X-axis at x={x}")
    case (0, y):
        print(f"On the Y-axis at y={y}")
    case (x, y):
        print(f"Point at ({x}, {y})")
```

## 5. Ternary Operator — One-Line Conditional

Syntax: `value_if_true if condition else value_if_false`

Use this for **simple** assignments or expressions. For complex logic, stick with full `if/else` blocks — readability matters.

```python
## Ternary operator — concise conditional expressions

age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age {age} → {status}")

# Practical uses
score = 85
result = "Pass" if score >= 60 else "Fail"
print(f"Score {score} → {result}")

# Inside an f-string
count = 1
print(f"You have {count} {'item' if count == 1 else 'items'} in your cart.")

# Nested ternary (use sparingly — can hurt readability)
x = 0
sign = "positive" if x > 0 else "negative" if x < 0 else "zero"
print(f"{x} is {sign}")
```

## Truthy and Falsy Values

Python does not require conditions to be strictly `True` or `False`. These values are considered **Falsy** (evaluate to `False`):

| Falsy Value | Type |
|---|---|
| `False` | bool |
| `0`, `0.0` | int, float |
| `""` (empty string) | str |
| `[]`, `()`, `{}` | list, tuple, dict |
| `None` | NoneType |

Everything else is **Truthy**.

```python
## Truthy and Falsy in action

name = ""
if name:
    print(f"Hello, {name}!")
else:
    print("Name is empty — please provide a name.")

items = []
if not items:
    print("Your cart is empty.")

# This is more Pythonic than:
# if len(items) == 0:  ← works but verbose
```

> **Best Practice:** Use Truthy/Falsy checks for emptiness (`if items:` instead of `if len(items) > 0`). It is more Pythonic and reads closer to natural English.
