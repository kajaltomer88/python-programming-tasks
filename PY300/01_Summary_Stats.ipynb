# PY300 — Summary Statistics
### Official documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html

Summary statistics are the **first thing you compute** when exploring a new dataset. They give you a quick sense of the data's shape, center, spread, and potential problems.

## What Are Summary Statistics?
A compact numerical description of a dataset — answering questions like:
- How many records? (`count`)
- What is the typical value? (`mean`, `median`)
- How spread out is the data? (`std`, `variance`, `range`, `IQR`)
- Are there extreme values? (`min`, `max`, percentiles)
- What is the shape of the distribution? (`skewness`, `kurtosis`)

## Approaches Covered
1. Pure Python — manual calculation
2. `statistics` module — standard library
3. NumPy — vectorized computation
4. Pandas `.describe()` — the go-to for real datasets

> **Perspective:** Summary statistics are your data's "vital signs". Just as a doctor checks pulse and temperature before diagnosis, a data analyst checks mean, median, and std before modelling. Never skip this step.

---
## Approach 1: Pure Python — From First Principles

Understanding what each statistic *actually computes* before relying on library shortcuts.

```python
## Manual summary statistics

data = [12, 15, 18, 22, 25, 28, 30, 35, 40, 100]

n = len(data)
total = sum(data)
mean = total / n

sorted_data = sorted(data)
mid = n // 2
median = (sorted_data[mid - 1] + sorted_data[mid]) / 2 if n % 2 == 0 else sorted_data[mid]

# Variance: average of squared differences from the mean
# Population variance uses n; Sample variance uses (n-1) — Bessel's correction
variance_pop = sum((x - mean) ** 2 for x in data) / n
variance_sample = sum((x - mean) ** 2 for x in data) / (n - 1)

std_pop = variance_pop ** 0.5
std_sample = variance_sample ** 0.5

data_range = max(data) - min(data)

# Interquartile Range (IQR) — range of the middle 50%
q1_idx = n // 4
q3_idx = 3 * n // 4
q1 = sorted_data[q1_idx]
q3 = sorted_data[q3_idx]
iqr = q3 - q1

print(f"Count    : {n}")
print(f"Mean     : {mean:.2f}")
print(f"Median   : {median}")
print(f"Min      : {min(data)}")
print(f"Max      : {max(data)}")
print(f"Range    : {data_range}")
print(f"Var (pop): {variance_pop:.2f}")
print(f"Var (sam): {variance_sample:.2f}")
print(f"Std (pop): {std_pop:.2f}")
print(f"Std (sam): {std_sample:.2f}")
print(f"Q1={q1}, Q3={q3}, IQR={iqr}")

# Learning Point: Sample variance divides by (n-1) instead of n.
# This is Bessel's correction — it compensates for the fact that a sample
# tends to underestimate the true population variance. Pandas and NumPy
# default to sample variance (ddof=1) for this reason.
```

## Approach 2: `statistics` Module — Standard Library

```python
## Summary stats using the statistics module

import statistics

data = [12, 15, 18, 22, 25, 28, 30, 35, 40, 100]

print(f"Mean       : {statistics.mean(data):.2f}")
print(f"Median     : {statistics.median(data)}")
print(f"Mode       : {statistics.mode(data)}")
print(f"Stdev (sam): {statistics.stdev(data):.2f}")      # Sample std (n-1)
print(f"Stdev (pop): {statistics.pstdev(data):.2f}")     # Population std (n)
print(f"Variance   : {statistics.variance(data):.2f}")

# Quantiles — divide data into equal-sized buckets
quartiles = statistics.quantiles(data, n=4)    # Q1, Q2 (median), Q3
print(f"Quartiles  : {quartiles}")

# Harmonic mean — useful for rates (speed, price-earnings ratios)
print(f"Harmonic mean: {statistics.harmonic_mean(data):.2f}")
```

## Approach 3: NumPy — Vectorized Computation

NumPy is 10–100x faster than pure Python for large arrays because it uses C-level operations on contiguous memory.

```python
## Summary stats using NumPy

import numpy as np

data = np.array([12, 15, 18, 22, 25, 28, 30, 35, 40, 100])

print(f"Count      : {data.size}")
print(f"Mean       : {np.mean(data):.2f}")
print(f"Median     : {np.median(data):.2f}")
print(f"Std (pop)  : {np.std(data):.2f}")              # Default: ddof=0 (population)
print(f"Std (sam)  : {np.std(data, ddof=1):.2f}")      # ddof=1 for sample
print(f"Var (pop)  : {np.var(data):.2f}")
print(f"Min        : {np.min(data)}")
print(f"Max        : {np.max(data)}")
print(f"Sum        : {np.sum(data)}")

# Percentiles
print(f"25th pctile: {np.percentile(data, 25):.2f}")
print(f"50th pctile: {np.percentile(data, 50):.2f}")   # Same as median
print(f"75th pctile: {np.percentile(data, 75):.2f}")
print(f"90th pctile: {np.percentile(data, 90):.2f}")

# Caution: NumPy defaults to ddof=0 (population), while Pandas defaults to
# ddof=1 (sample). This is a common source of confusion when comparing results.
```

## Approach 4: Pandas `.describe()` — The EDA Workhorse

For real datasets, Pandas gives you a complete summary in one call.

```python
## Summary stats using Pandas

import pandas as pd

# Create a sample dataset
df = pd.DataFrame({
    'age':    [25, 30, 35, 40, 45, 28, 33, 50, 22, 60],
    'salary': [30000, 45000, 55000, 60000, 80000, 40000, 52000, 95000, 28000, 120000],
    'rating': [4.2, 3.8, 4.5, 4.0, 4.8, 3.5, 4.1, 4.9, 3.2, 4.7],
    'dept':   ['IT', 'HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance']
})

# .describe() — summary for numeric columns
print("=== Numeric Summary ===")
print(df.describe())

print("\n=== Include All Columns (categorical too) ===")
print(df.describe(include='all'))

print("\n=== Custom Percentiles ===")
print(df.describe(percentiles=[0.1, 0.25, 0.5, 0.75, 0.9]))
```

```python
## Individual Pandas stat methods

print(f"Mean salary    : {df['salary'].mean():,.2f}")
print(f"Median salary  : {df['salary'].median():,.2f}")
print(f"Std salary     : {df['salary'].std():,.2f}")     # ddof=1 by default
print(f"Skewness       : {df['salary'].skew():.2f}")     # >0 = right-skewed
print(f"Kurtosis       : {df['salary'].kurtosis():.2f}") # >0 = heavy tails

# Group-level summary
print("\n=== Salary Stats by Department ===")
print(df.groupby('dept')['salary'].describe())

# Correlation matrix — how strongly are numeric columns related?
print("\n=== Correlation Matrix ===")
print(df[['age', 'salary', 'rating']].corr().round(2))
```

---
## Skewness and Kurtosis — Shape of the Distribution

| Measure | What It Tells You | Value Interpretation |
|---|---|---|
| **Skewness** | Symmetry of the distribution | 0 = symmetric, >0 = right tail, <0 = left tail |
| **Kurtosis** | Heaviness of the tails | 0 = normal-like, >0 = heavy tails (outliers), <0 = light tails |

> **When to care:** Skewness > 1 or < -1 often means the mean is misleading — use the median instead. High kurtosis warns you that outliers are likely.

## Quick Reference

| Statistic | Pure Python | `statistics` | NumPy | Pandas |
|---|---|---|---|---|
| Mean | `sum(d)/len(d)` | `mean(d)` | `np.mean(d)` | `df.mean()` |
| Median | manual sort | `median(d)` | `np.median(d)` | `df.median()` |
| Std (sample) | manual | `stdev(d)` | `np.std(d, ddof=1)` | `df.std()` |
| Variance | manual | `variance(d)` | `np.var(d, ddof=1)` | `df.var()` |
| Percentile | manual | `quantiles(d)` | `np.percentile(d, q)` | `df.quantile(q)` |
| Full summary | — | — | — | `df.describe()` |

> **Best Practice:** Start every EDA with `df.describe()` + `df.info()` + `df.isnull().sum()`. These three calls tell you the shape, types, stats, and data quality in under 5 seconds.
