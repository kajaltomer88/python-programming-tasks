# PY300 — Data Visualizations — Charts and Graphs
### Official documentation: https://matplotlib.org/stable/gallery/index.html

Visualization turns numbers into **stories**. Choosing the right chart type is as important as computing the right statistic.

## Chart Types Covered
1. **Bar Chart** — comparing categories
2. **Line Graph** — trends over time
3. **Histogram** — distribution of a single variable
4. **Pie Chart** — composition / proportion
5. **Stacked Bar/Area Chart** — composition over categories/time
6. **Scatter Plot** — relationship between two variables
7. **Heatmap** — correlation or matrix data
8. **Waterfall Chart** — cumulative contribution

## Other Visualization Types (Perspective)
| Chart | When to Use |
|---|---|
| **Violin Plot** | Distribution + density (richer than box plot) |
| **Treemap** | Hierarchical composition (better than nested pies) |
| **Sunburst** | Multi-level hierarchical data |
| **Sankey Diagram** | Flow/transition between stages |
| **Radar/Spider Chart** | Multi-dimensional comparison of entities |
| **Bubble Chart** | Scatter + third variable as bubble size |
| **Pair Plot** | All pairwise relationships (seaborn) |
| **Funnel Chart** | Conversion/drop-off analysis |

> **Perspective:** The best visualization is the one your audience understands in under 5 seconds. Avoid chart junk, 3D effects, and dual axes unless truly necessary. Edward Tufte's principle: *maximize the data-ink ratio*.

```python
## Setup — sample data we will reuse

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Monthly sales data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2024 = [120, 135, 150, 170, 160, 180, 200, 210, 190, 220, 250, 280]
sales_2025 = [130, 145, 165, 175, 185, 195, 215, 225, 205, 240, 270, 310]

# Category data
departments = ['IT', 'HR', 'Finance', 'Marketing', 'Operations']
headcount = [45, 20, 15, 30, 40]
budget = [500, 150, 200, 350, 400]  # in thousands
```

---
## 1. Bar Chart — Comparing Categories

Use when you have **categorical data** and want to compare values across groups.

```python
## Bar chart — vertical and horizontal

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Vertical bar
colors = ['#2196F3', '#4CAF50', '#FF9800', '#E91E63', '#9C27B0']
axes[0].bar(departments, headcount, color=colors)
axes[0].set_title('Headcount by Department')
axes[0].set_ylabel('Employees')

# Horizontal bar — better when labels are long
axes[1].barh(departments, budget, color=colors)
axes[1].set_title('Budget by Department (₹ thousands)')
axes[1].set_xlabel('Budget (₹K)')

plt.tight_layout()
plt.show()

# Tip: Use horizontal bars when category labels are long.
# Always sort bars by value (not alphabetically) for easier comparison.
```

---
## 2. Line Graph — Trends Over Time

```python
## Line graph — time series comparison

plt.figure(figsize=(12, 5))
plt.plot(months, sales_2024, marker='o', label='2024', linewidth=2)
plt.plot(months, sales_2025, marker='s', label='2025', linewidth=2, linestyle='--')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (units)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Best Practice: Line charts imply continuity — only use for ordered data
# (time, sequence). For unordered categories, use bar charts instead.
```

---
## 3. Histogram — Distribution of a Variable

```python
## Histogram — see the shape of your data

np.random.seed(42)
salaries = np.concatenate([
    np.random.normal(50000, 12000, 300),   # Main cluster
    np.random.normal(100000, 8000, 50)      # High earners
])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Basic histogram
axes[0].hist(salaries, bins=30, color='steelblue', edgecolor='white')
axes[0].set_title('Salary Distribution')
axes[0].set_xlabel('Salary (₹)')
axes[0].set_ylabel('Count')
axes[0].axvline(np.mean(salaries), color='red', linestyle='--', label=f'Mean: {np.mean(salaries):,.0f}')
axes[0].axvline(np.median(salaries), color='green', linestyle='--', label=f'Median: {np.median(salaries):,.0f}')
axes[0].legend()

# KDE overlay (smooth density curve)
axes[1].hist(salaries, bins=30, density=True, color='lightblue', edgecolor='white')
from scipy.stats import gaussian_kde
kde = gaussian_kde(salaries)
x_range = np.linspace(salaries.min(), salaries.max(), 200)
axes[1].plot(x_range, kde(x_range), color='darkblue', linewidth=2)
axes[1].set_title('Salary Distribution with KDE')
axes[1].set_xlabel('Salary (₹)')

plt.tight_layout()
plt.show()

# Notice: Mean > Median signals RIGHT skew (pulled by high earners).
# This is why median salary is more representative than mean salary.
```

---
## 4. Pie Chart — Composition / Proportion

```python
## Pie chart

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Basic pie
axes[0].pie(headcount, labels=departments, autopct='%1.1f%%',
            colors=colors, startangle=90)
axes[0].set_title('Headcount Distribution')

# Donut chart (pie with a hole) — often cleaner
wedges, texts, autotexts = axes[1].pie(
    headcount, labels=departments, autopct='%1.1f%%',
    colors=colors, startangle=90, pctdistance=0.82)
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
axes[1].add_artist(centre_circle)
axes[1].set_title('Headcount — Donut Chart')

plt.tight_layout()
plt.show()

# Perspective: Pie charts are controversial. Humans are bad at comparing
# angles. For more than 5 categories, use a horizontal bar chart instead.
# Pie charts work best when one slice clearly dominates.
```

---
## 5. Stacked Bar / Stacked Area — Composition Over Categories or Time

```python
## Stacked bar chart

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Stacked bar
x = np.arange(len(months))
axes[0].bar(x, sales_2024, label='2024', color='steelblue')
axes[0].bar(x, sales_2025, bottom=sales_2024, label='2025', color='coral')
axes[0].set_xticks(x)
axes[0].set_xticklabels(months, rotation=45)
axes[0].set_title('Stacked Bar — Combined Sales')
axes[0].legend()

# Stacked area
axes[1].stackplot(months, sales_2024, sales_2025,
                  labels=['2024', '2025'], colors=['steelblue', 'coral'], alpha=0.7)
axes[1].set_title('Stacked Area — Sales Over Time')
axes[1].legend(loc='upper left')
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

---
## 6. Scatter Plot — Relationship Between Two Variables

```python
## Scatter plot — with regression line

np.random.seed(42)
experience = np.random.uniform(1, 20, 50)
salary = 25000 + 4000 * experience + np.random.normal(0, 8000, 50)

plt.figure(figsize=(10, 6))
plt.scatter(experience, salary, alpha=0.7, c='steelblue', edgecolors='white', s=80)

# Add trend line
z = np.polyfit(experience, salary, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 20, 100)
plt.plot(x_line, p(x_line), color='red', linewidth=2, linestyle='--', label='Trend line')

plt.title('Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (₹)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

---
## 7. Heatmap — Correlation Matrix

```python
## Heatmap — correlation matrix

import seaborn as sns

np.random.seed(42)
df = pd.DataFrame({
    'Experience': np.random.uniform(1, 20, 100),
    'Salary': np.random.uniform(30000, 120000, 100),
    'Rating': np.random.uniform(2.0, 5.0, 100),
    'Projects': np.random.randint(1, 30, 100)
})
df['Salary'] = 25000 + 4000 * df['Experience'] + np.random.normal(0, 5000, 100)

corr = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

# Reading a heatmap: values close to +1 = strong positive correlation,
# close to -1 = strong negative, close to 0 = no linear relationship.
```

## 8. Waterfall Chart — Cumulative Contribution

```python
## Waterfall chart — showing how components add up to a total

categories = ['Revenue', 'COGS', 'Gross Profit', 'OpEx', 'Tax', 'Net Profit']
values =     [1000,      -400,    0,              -250,   -100,  0]

# Calculate running total
running = []
total = 0
for i, v in enumerate(values):
    if categories[i] in ('Gross Profit', 'Net Profit'):  # Subtotal rows
        running.append(total)
    else:
        total += v
        running.append(total)

# Build the waterfall
fig, ax = plt.subplots(figsize=(12, 6))
bottoms = [0] * len(categories)
color_list = []
display_vals = []

prev = 0
for i, (cat, val) in enumerate(zip(categories, values)):
    if cat in ('Gross Profit', 'Net Profit'):
        bottoms[i] = 0
        display_vals.append(running[i])
        color_list.append('#2196F3')     # Blue for totals
    elif val >= 0:
        bottoms[i] = prev
        display_vals.append(val)
        color_list.append('#4CAF50')     # Green for positive
    else:
        bottoms[i] = prev + val
        display_vals.append(abs(val))
        color_list.append('#F44336')     # Red for negative
    prev = running[i]

ax.bar(categories, display_vals, bottom=bottoms, color=color_list, edgecolor='white', width=0.6)

for i, (b, v) in enumerate(zip(bottoms, display_vals)):
    ax.text(i, b + v + 10, f'{running[i]}', ha='center', fontweight='bold')

ax.set_title('Profit Waterfall Chart')
ax.set_ylabel('Amount (₹K)')
ax.axhline(y=0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

## Chart Selection Guide

| Question You Are Asking | Chart Type |
|---|---|
| How do categories compare? | Bar chart |
| How does a value change over time? | Line graph |
| What does the distribution look like? | Histogram, violin plot |
| What is the proportion/composition? | Pie, donut, treemap |
| How do parts add up to a whole over time? | Stacked bar/area |
| Is there a relationship between two variables? | Scatter plot |
| How are all variables correlated? | Heatmap |
| How do increments/decrements reach a total? | Waterfall |
| How does data flow between stages? | Sankey diagram |

> **Best Practice:** Always label your axes, include units, and add a descriptive title. Use colour intentionally — not decoratively. A chart should be interpretable without reading the surrounding text.
