# PY200 — Finding the Largest and Smallest Numbers

A classic beginner exercise that reinforces **loops**, **comparisons**, and **variable tracking**.

## What We Will Print
Given a list of numbers, find and display:
1. The **largest** number
2. The **second largest** number
3. The **smallest** number

> **Why this matters:** This pattern — scanning a collection while tracking state — appears everywhere: finding the best score, the cheapest product, the most recent date, etc.

---
## Approach 1: Using Built-in Functions (`max`, `min`, `sorted`)

Python provides built-in functions that make this trivial. **Always know the easy way first** before implementing from scratch.

```python
## Built-in approach — the Pythonic way

numbers = [42, 17, 93, 55, 8, 76, 31, 93, 64]

largest = max(numbers)
smallest = min(numbers)

# For second largest: remove duplicates with set(), sort descending, pick index 1
unique_sorted = sorted(set(numbers), reverse=True)
second_largest = unique_sorted[1] if len(unique_sorted) > 1 else unique_sorted[0]

print(f"Largest        : {largest}")
print(f"Second Largest : {second_largest}")
print(f"Smallest       : {smallest}")

# Time Complexity: O(n) for max/min, O(n log n) for sorted
# Best Practice: For just largest/smallest, max()/min() are O(n) and sufficient.
# sorted() is overkill if you only need top 2 — but it is readable and correct.
```

---
## Approach 2: Manual Loop — Track Largest, Second Largest, and Smallest

This is the **first-principles** approach. Walk through the list once, keeping track of three variables.

**Time Complexity:** O(n) — single pass through the list. This is optimal.

```python
## Manual single-pass approach — O(n)

numbers = [42, 17, 93, 55, 8, 76, 31, 93, 64]

# Initialize with negative/positive infinity so any real number will replace them
largest = float('-inf')
second_largest = float('-inf')
smallest = float('inf')

for num in numbers:
    # Update smallest
    if num < smallest:
        smallest = num

    # Update largest and second largest
    if num > largest:
        second_largest = largest    # Demote current largest to second
        largest = num               # New largest found
    elif num > second_largest and num != largest:
        second_largest = num        # New second largest (must be different from largest)

print(f"Largest        : {largest}")
print(f"Second Largest : {second_largest}")
print(f"Smallest       : {smallest}")

# Learning Point: We use float('-inf') and float('inf') as initial values.
# This avoids the common beginner mistake of initializing to 0 (which fails
# if all numbers are negative or all are positive and > 0).
```

---
## Approach 3: Using `heapq.nlargest` / `heapq.nsmallest`

When you need the **top N** or **bottom N** elements, the `heapq` module is the optimal choice.

**Time Complexity:** O(n log k) where k is the number of elements you want. Better than full sorting when k << n.

```python
## heapq approach — best for "top N" problems

import heapq

numbers = [42, 17, 93, 55, 8, 76, 31, 93, 64]

top_2 = heapq.nlargest(2, set(numbers))   # Unique top 2
bottom_1 = heapq.nsmallest(1, numbers)

print(f"Largest        : {top_2[0]}")
print(f"Second Largest : {top_2[1]}")
print(f"Smallest       : {bottom_1[0]}")

# Best Practice: Use heapq when you need top/bottom N from a large dataset.
# For just 1 element, max()/min() is simpler and equally fast.
# For top 2-3 from millions of records, heapq shines.
```

---
## Summary: Which Approach to Use?

| Approach | Time Complexity | Best When |
|---|---|---|
| `max()` / `min()` | O(n) | You need only the single largest or smallest |
| Manual loop | O(n) | You need top 2–3 and want a single pass with no extra memory |
| `sorted()` | O(n log n) | You need the full ranking or the list is small |
| `heapq.nlargest()` | O(n log k) | You need top/bottom k from a large dataset |

> **Perspective:** In interviews and real code, knowing *which tool fits the problem size* matters more than always writing the fastest solution. Start readable, optimize when profiling tells you to.
