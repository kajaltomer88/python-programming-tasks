# PY200 — Calculating the Mode

The **mode** is the value that appears **most frequently** in a dataset. Unlike the mean and median, the mode can apply to non-numeric data too (e.g., most popular color, most common word).

## Key Facts
- A dataset can have **no mode** (all values appear equally), **one mode** (unimodal), or **multiple modes** (bimodal, multimodal).
- The mode is the only measure of central tendency that works with **categorical data**.

## Implementations Covered
1. **Two nested `for` loops** — brute force, first principles
2. **Single loop with `in` operator** — cleaner manual approach
3. **`Counter` from `collections`** — the Pythonic way

> **Perspective:** Finding the mode is really a **frequency counting** problem. This pattern appears everywhere: most visited page, most sold product, most common error code, etc.

---
## Approach 1: Two Nested `for` Loops — Brute Force

For each element, loop through the entire list to count how many times it appears. Track the element with the highest count.

**Time Complexity:** O(n²) — for each of `n` elements, we scan all `n` elements again.  
**Space Complexity:** O(1) — no extra data structures.

```python
## Mode using two nested for loops — O(n²)

data = [3, 5, 5, 5, 7, 9, 7, 3, 5, 7]

mode_value = data[0]       # Assume first element is the mode
max_count = 0              # Highest frequency seen so far

# Outer loop: pick each element as a candidate
for candidate in data:
    count = 0
    # Inner loop: count how many times 'candidate' appears in the full list
    for element in data:
        if element == candidate:
            count += 1

    # Update mode if this candidate appears more often
    if count > max_count:
        max_count = count
        mode_value = candidate

print(f"Data : {data}")
print(f"Mode : {mode_value} (appears {max_count} times)")

# Learning Point: This works but is SLOW for large datasets.
# For a list of 10,000 elements, the inner loop runs 10,000 x 10,000
# = 100,000,000 times. That's 100 million comparisons!
#
# Also, this approach recounts duplicates — if 5 appears 4 times,
# we count it 4 separate times (once for each occurrence). Wasteful.
```

---
## Approach 2: Single Loop with `in` Operator — Build a Frequency Map

Instead of recounting every element, we build a **dictionary** that maps each unique value to its frequency. Then we find the key with the highest value.

**Time Complexity:** O(n) — single pass to build the frequency dictionary.  
**Space Complexity:** O(k) — where k is the number of unique elements.

```python
## Mode using a dictionary and the `in` operator — O(n)

data = [3, 5, 5, 5, 7, 9, 7, 3, 5, 7]

frequency = {}               # Empty dictionary to store value → count

for value in data:
    if value in frequency:   # 'in' checks if key exists — O(1) average for dicts
        frequency[value] += 1
    else:
        frequency[value] = 1

print(f"Frequency map: {frequency}")

# Find the key with the highest count
mode_value = None
max_count = 0

for value, count in frequency.items():
    if count > max_count:
        max_count = count
        mode_value = value

print(f"Mode : {mode_value} (appears {max_count} times)")

# Learning Point: The 'in' operator on a dictionary is O(1) on average
# because dictionaries use hash tables. This is why the overall
# algorithm is O(n) — a huge improvement over the O(n²) nested loops.
```

```python
## Cleaner version using dict.get() — avoids the if/else check

data = [3, 5, 5, 5, 7, 9, 7, 3, 5, 7]

frequency = {}
for value in data:
    # .get(key, default) returns the current count or 0 if key doesn't exist
    frequency[value] = frequency.get(value, 0) + 1

# Use max() with key= to find the mode in one line
mode_value = max(frequency, key=frequency.get)

print(f"Frequency: {frequency}")
print(f"Mode: {mode_value} (appears {frequency[mode_value]} times)")

# Best Practice: dict.get(key, default) is more Pythonic than
# checking 'if key in dict'. It reads as one clear intention.
```

---
## Approach 3: Using `Counter` from `collections` — The Pythonic Way

`Counter` is a specialized dictionary subclass designed exactly for counting. It provides `.most_common()` which returns elements sorted by frequency.

**Time Complexity:** O(n) to build + O(k log k) for `.most_common()` where k = unique elements.  
**Space Complexity:** O(k)

```python
## Mode using Counter — the professional approach

from collections import Counter

data = [3, 5, 5, 5, 7, 9, 7, 3, 5, 7]

# Counter builds the frequency map automatically
freq = Counter(data)
print(f"Counter  : {freq}")
print(f"As dict  : {dict(freq)}")

# .most_common(n) returns the n most frequent elements as (value, count) tuples
print(f"Top 3    : {freq.most_common(3)}")

# The mode is simply the first element of most_common(1)
mode_value, mode_count = freq.most_common(1)[0]
print(f"Mode     : {mode_value} (appears {mode_count} times)")
```

```python
## Handling multiple modes (multimodal data)

from collections import Counter

data = [1, 2, 2, 3, 3, 4]   # Both 2 and 3 appear twice — bimodal

freq = Counter(data)
max_count = freq.most_common(1)[0][1]    # Get the highest frequency

# Filter all values that have the max frequency
modes = [value for value, count in freq.items() if count == max_count]

if len(modes) == len(freq):
    print(f"No mode — all values appear equally ({max_count} times each)")
elif len(modes) == 1:
    print(f"Unimodal: {modes[0]} (appears {max_count} times)")
else:
    print(f"Multimodal: {modes} (each appears {max_count} times)")
```

```python
## Bonus: Using statistics.mode() and statistics.multimode()

import statistics

data_single = [3, 5, 5, 5, 7, 9, 7, 3, 5, 7]
data_multi = [1, 2, 2, 3, 3, 4]

print(f"statistics.mode()      : {statistics.mode(data_single)}")

# multimode() returns ALL modes (Python 3.8+)
print(f"statistics.multimode() : {statistics.multimode(data_multi)}")

# Works with strings too!
colors = ["red", "blue", "red", "green", "blue", "red"]
print(f"Most common color     : {statistics.mode(colors)}")
```

---
## Comparison of All Approaches

| Approach | Time | Space | Handles Multimodal? | Readability |
|---|---|---|---|---|
| Nested `for` loops | O(n²) | O(1) | No | Low — hard to extend |
| Dict with `in` | O(n) | O(k) | With extra logic | Medium |
| `Counter.most_common()` | O(n + k log k) | O(k) | Easily | High |
| `statistics.mode()` | O(n) | O(k) | Use `multimode()` | Highest |

### Optimal Perspective

**For production code**, use `Counter` — it is battle-tested, expressive, and versatile. You can do far more than just mode: top-N elements, frequency histograms, set-like arithmetic (`Counter + Counter`), etc.

**For interviews**, demonstrate the dictionary approach first (shows you understand hash maps), then mention `Counter` as the Pythonic shortcut. Knowing both signals depth.

**Avoid the O(n²) approach** unless you are explicitly asked to solve it without any data structures. It does not scale.

> **Key Takeaway:** The jump from O(n²) to O(n) comes from using a **hash map** (dictionary). This is one of the most important patterns in programming — *trade space for time*.
